Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

Load baseline + HPO results

```python
baseline = pd.read_csv("../results/lr_creditg_baseline.csv")
hpo = pd.read_csv("../results/lr_creditg_hpo.csv")

baseline_f1 = baseline['baseline_f1_macro'].iloc[0]
hpo_f1 = hpo['hpo_f1_macro'].iloc[0]
```

Plot

```python
plt.figure(figsize=(6,4))

plt.bar(["Baseline LR", "HPO LR"], [baseline_f1, hpo_f1])
plt.ylabel("F1 Macro")
plt.title("Logistic Regression on credit-g: Baseline vs HPO")

os.makedirs("../figures", exist_ok=True)
plt.savefig("../figures/lr_creditg_baseline_vs_hpo.png", dpi=200, bbox_inches="tight")

plt.show()
```
