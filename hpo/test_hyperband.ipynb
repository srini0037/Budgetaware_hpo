Load Small Test Data

```python
# Quick test with small dataset first
from sklearn.datasets import make_classification

print("Creating small test dataset...")
X_test, y_test = make_classification(
    n_samples=1000, 
    n_features=20,
    n_informative=10,  # Added this
    n_redundant=5,      # Added this
    n_classes=3, 
    n_clusters_per_class=2,
    random_state=42
)

print(f"Test data shape: {X_test.shape}")
print(f"Classes: {len(np.unique(y_test))}")
print("✅ Test data created!")
```

Test Hyperband (Quick)

```python
print("Testing Hyperband on small dataset...")
print("=" * 70)

hb = Hyperband(
    get_random_config=get_random_mlp_config,
    max_iter=27,  # Small for quick test
    eta=3,
    verbose=True
)

result = hb.run(X_test, y_test)

print("\n" + "=" * 70)
print("✅ HYPERBAND TEST COMPLETE!")
print("=" * 70)
print(f"Best score: {result['best_score']:.4f}")
print(f"Configs evaluated: {result['configs_evaluated']}")
print(f"Total time: {result['total_time']:.2f}s")
```
